```python
# Dependencies
import pandas as pd
import json
```

```python
# Import csv file as dataframe
df1 = pd.read_csv("Data/athlete_events.csv")
df1 = df1[df1['Season']=='Summer']
```

```python
df2 = pd.read_csv("Data/noc_regions.csv")
df2.head()
```

```python
# Merge df2 to df1
df = df1.merge(df2,how="left",on="NOC")
```

```python
# Count Athelete numbers 
df_athelete =  df.groupby(['Year',"Sex","Sport"]).Name.count().reset_index()
df_athelete = df_athelete.rename(columns={'Name':"Athelete_count"}).fillna(0)
df_athelete.head()
```

```python
df_athelete.Sport.unique()  
```

```python
df_athelete.info()
```

```python
df_athelete.Year.unique()
```

```python
# Save it as csv file
df_athelete.to_csv("Data/athelete_count.csv",index=False)
```

```python
# Save dataframe as json file
data_athelete = df_athelete.to_json(orient='records')
```

```python
data_athelete
```

```python
# Convert str to dictionary
data_athelete = json.loads(data_athelete)
```

```python
# Save dictionary as json file
with open('Data/athelete.json', 'w') as fp:
    json.dump(data_athelete, fp)
```

```python
# Medal count for each nation each game 

df3 = df[['NOC', 'Games','Year','Event', 'Medal', 'region']].drop_duplicates().dropna(subset=['Medal'])
```

```python
df3.shape
```

```python
df3.head()
```

```python
df4 = pd.DataFrame(df3.groupby(['Year','NOC']).Medal.count())
df4 = df4.rename(columns={"Medal":"Medal_count"})
df4 = df4.reset_index()
df4.head()
```

```python
df5 = pd.DataFrame(df3[df3["Medal"]=="Gold"].groupby(['Year','NOC']).Medal.count())
df5 = df5.rename(columns={"Medal":"Gold_count"})
df5 = df5.reset_index()
df5.head()
```

```python
df6 = pd.DataFrame(df3[df3["Medal"]=='Silver'].groupby(['Year','NOC']).Medal.count())
df6 = df6.rename(columns={"Medal":"Silver_count"})
df6 = df6.reset_index()
df6.head()
```

```python
df7 = pd.DataFrame(df3[df3["Medal"]=='Bronze'].groupby(['Year','NOC']).Medal.count())
df7 = df7.rename(columns={"Medal":"Bronze_count"})
df7= df7.reset_index()
df7.head()
```

```python
# Combine dataframe 
df_medal = df4.merge(df5,how="left",on=['Year','NOC'])
df_medal = df_medal.merge(df6,how="left",on=['Year','NOC'])
df_medal = df_medal.merge(df7,how="left",on=['Year','NOC'])
df_medal = df_medal.fillna(0)
df_medal.head()
```

```python
df_medal.info()
```

```python
# Convert float to int
df_medal['Gold_count']=df_medal['Gold_count'].apply(lambda x: int(x))
df_medal['Silver_count']=df_medal['Silver_count'].apply(lambda x: int(x))
df_medal['Bronze_count']=df_medal['Bronze_count'].apply(lambda x: int(x))
```

```python
df_medal.info()
```
